```python
# Hacer policy_bind_date a datetime
df["policy_bind_date"] = pd.to_datetime(df["policy_bind_date"])
df["incident_date"] = pd.to_datetime(df["incident_date"])
print("policy_bind_date: ", df["policy_bind_date"].dtype)
print("incident_date: ", df["incident_date"].dtype)
```

```python
df.dtypes[20:40]
```

```python
# Sacar todas las columnas que tengan "?"
print(df.columns[df.isin(["?"]).any()])
print(df.columns[df.isin(["YES"]).any()])
print(df.columns[df.isin(["Y"]).any()])
print(df.columns[df.isin(["NO"]).any()])
print(df.columns[df.isin(["N"]).any()])
```

```python
# Uniques de ['collision_type', 'property_damage', 'police_report_available']
print(df["police_report_available"].unique())
print(df["collision_type"].unique())
print(df["property_damage"].unique())
print(df["fraud_reported"].unique())

print(df["police_report_available"].value_counts())
print(df["property_damage"].value_counts())
print(df["fraud_reported"].value_counts())
```

```python
# Mapear los valores "?" a "NO" en las columnas 'property_damage' y 'police_report_available'
cols_con_pregunta = ['property_damage', 'police_report_available']
for col in cols_con_pregunta:
    df[col] = df[col].replace('?', 'NO')
    
df["fraud_reported"] = df["fraud_reported"].replace('N', 'NO')
df["fraud_reported"] = df["fraud_reported"].replace('Y', 'YES')

print(df["police_report_available"].unique())
print(df["property_damage"].unique())
print(df["fraud_reported"].unique())
# count cuantas veces aparece "NO" en la columna y cuantas veces aparece "?" en la columna 'police_report_available' 
print(df["police_report_available"].value_counts())
print(df["property_damage"].value_counts())
print(df["fraud_reported"].value_counts())
```
